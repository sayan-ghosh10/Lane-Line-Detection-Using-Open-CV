# Canny Edge Detection

Wikipedia: The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images.

# Gray scaling the images

The Canny edge detection algorithm measures the intensity gradients of each pixel. So, we need to convert the images into gray scale in order to detect edges.

# Applying Gaussian smoothing

Since all edge detection results are easily affected by image noise, it is essential to filter out the noise to prevent false detection caused by noise. To smooth the image, a Gaussian filter is applied to convolve with the image. This step will slightly smooth the image to reduce the effects of obvious noise on the edge detector.

# Applying Canny Edge Detection

The Process of Canny edge detection algorithm can be broken down to 5 different steps:
1. Find the intensity gradients of the image
2. Apply non-maximum suppression to get rid of spurious response to edge detection.
3. Apply *double threshold* to determine potential edges.
4. Track edge by hysteresis: Finalize the detection of edges by suppressing all the other edges that are weak and not connected to strong edges.

*If an edge pixel’s gradient value is higher than the high threshold value, it is marked as a strong edge pixel. If an edge pixel’s gradient value is smaller than the high threshold value and larger than the low threshold value, it is marked as a weak edge pixel. If an edge pixel's value is smaller than the low threshold value, it will be suppressed.
The two threshold values are empirically determined and their definition will depend on the content of a given input image.*

```python
def canny(image):
    gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    blur=cv2.GaussianBlur(gray,(5,5),0)
    canny=cv2.Canny(blur,50,150)
    return canny
```

# Region of interest

We're interested in the area facing the camera, where the lane lines are found. So, we'll apply region masking to cut out everything else.

```python
def outline(image):
    height=image.shape[0]
    polygon=np.array([[
        (20,height),(940,height),(500,317)]
    ])
    mask=np.zeros_like(image)
    cv2.fillPoly(mask,polygon,255)
    masked_image=cv2.bitwise_and(image,mask)
    return masked_image
```

```python
def display_lines(lines,image):
        line_image=np.zeros_like(image)
        img=np.copy(line_image)
        if lines is not None:
            for x1,y1,x2,y2 in lines:
                cv2.line(img,(x1,y1),(x2,y2),(255,0,0),10)
        return line_image
```

```python
def coordinates(image,line_parameters):
    try:
        slope,intercept=line_parameters
    except TypeError:
        slope, intercept = 0.001, 0
    y1=image.shape[0]
    y2=int(y1*(3/5))
    x1=int((y1-intercept)/slope)
    x2=int((y2-intercept)/slope)
    return np.array([x1,y1,x2,y2])
```

## Average slope and intercept of the lane lines
We have multiple lines detected for each lane line. We need to average all these lines and draw a single line for each lane line.
We also need to extrapolate the lane lines to cover the full lane line length.

```python
def average_slope_intercept(image,lines):
    left_fit=[]
    right_fit=[]
    for line in lines:
        x1,y1,x2,y2=line.reshape(4)
        parameters=np.polyfit((x1,x2),(y1,y2),1)
        slope=parameters[0]
        intercept=parameters[1]
        if slope<0:
            left_fit.append((slope,intercept))
        else:
            right_fit.append((slope,intercept))
    avg_left=np.average(left_fit,axis=0)
    avg_right=np.average(right_fit,axis=0)
    left_line=coordinates(image,avg_left)
    right_line=coordinates(image,avg_right)
    return np.array([left_line,right_line])
```

# Lane Line Detection in  Images:

```python
def list_images(images, cols = 2, rows = 5, cmap=None):
    plt.figure(figsize=(10, 11))
    for i, image in enumerate(images):
        plt.subplot(rows, cols, i+1)
        cmap = 'gray' if len(image.shape) == 2 else cmap
        plt.imshow(image, cmap = cmap)
        plt.xticks([])
        plt.yticks([])
    plt.tight_layout(pad=0, h_pad=0, w_pad=0)
    plt.show()
```

```python
image=cv2.imread('7.jpg')
lane_image=np.copy(image)
canny_image=canny(lane_image)
outline_image=outline(canny_image)
lines=cv2.HoughLinesP(outline_image,2, np.pi/180,100, np.array([]) , minLineLength=40,maxLineGap=5)
average_lines=average_slope_intercept(lane_image,lines)
line_image=display_lines(average_lines,lane_image)
combo=cv2.addWeighted(lane_image,0.8,line_image,1,1)
cv2.imshow('result',combo)
cv2.waitKey(0)
```


# Hough Transform

The Hough transform is a technique which can be used to isolate features of a particular shape within an image. We'll use it to detected the lane lines in `outline()`.

## Apply on video streams 

Now, we'll use the above functions to detect lane lines from a video stream.

```python
cap=cv2.VideoCapture('2vid.mp4')
while (cap.isOpened()):
    _,frame=cap.read()
    if frame is not None:
        canny_image=canny(frame)
        img=outline(canny_image)
        lines=cv2.HoughLinesP(img ,2, np.pi/180,100, np.array([]) , minLineLength=40,maxLineGap=5)
        average_lines=average_slope_intercept(frame,lines)
        line_image=display_lines(average_lines,frame)
        combo=cv2.addWeighted(frame,0.8,line_image,1,1)
        cv2.imshow('result',combo)
    if cv2.waitKey(1) & 0xFF ==ord('b'):
        break
cap.release()
cv2.destroyAllWindows()
```
